```python
from kokoro import KPipeline
import soundfile as sf
import numpy as np

# Define o código de linguagem para português
lang_code = "p"

# Inicializa o pipeline com o código de linguagem desejado
pipeline = KPipeline(lang_code=lang_code)


# Texto que será convertido em fala
text = '''
Voc~es não precisam mais pagar por geradores de vozes, pessoal...
porque agora podem gerar vozes como a minha, a de papai noel
'''

generator = pipeline(text, voice='pm_santa')

audio_chunks = []

for gs, ps, audio in generator:
    # gs: Global state
    # ps: Pipeline state
    # audio: Audio chunk
    audio_chunks.append(audio)

audio_completo = np.concatenate(audio_chunks)
sf.write('audio_completo.wav', audio_completo, 24000)
```
